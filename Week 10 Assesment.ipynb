1.1 Problem 1: Tip Calculator
1.1.1 The Scenario
You work at a Perth restaurant. The manager wants a simple tool where staff can enter:

The bill amount (e.g., 45.80)
The tip percentage (e.g., 15%)
And get back:

The tip amount
The total amount to charge
1.1.2 What You Must Submit
STEP 1: Understand the Problem

In your own words, what is this program supposed to do? (2-3 sentences)

This program is designed to help restaurant staff quickly calculate a customer’s tip and final bill total. The user enters the bill amount and the tip percentage, and the program works out how much the tip is and the total amount the customer needs to pay.

STEP 2: Identify Inputs and Outputs

INPUTS:The customer’s bill amount entered by the staff member
Example: 45.80
Data type: Decimal number because the amount can include cents
The tip percentage entered by the staff member
Example: 15%
Data type: Number/percentage


[What does the user provide?]
The user provides two inputs: the bill amount and the tip percentage.The bill amount is basically the total cost of the customer’s food and drinks before adding the tip. ,and the tip percentage is the percentage of extra money the customer wants to give as a tip for the service.

[What type of data? Numbers? Text?]
The input data are numbers (decimal and percentage values).
OUTPUTS:The tip amount calculated from the bill and tip percentage
Example: 6.87
Displayed in dollar format with two decimal places
The total amount the customer needs to pay after the tip is added
Example: 52.67
Displayed in dollar format with two decimal places so staff can charge the correct total amount

Displayed as a dollar amount

Total amount to charge
[What does the program show back?]
The program shows back the tip amount in dollars and the final total amount the customer must pay after the tip has been added to the original bill. 

[What format?]
The results should be displayed as currency values (for example: (6.87) )
  meaning it should be displayed in dollar format $ usually with 2 decimal places.
STEP 3: Work It By Hand

Show me you can do this with a calculator. Use this example:

Bill amount: 45.80
Tip percentage: 15%
Write the steps:

[First, convert 15% into a decimal:
15% = 0.15]
[Multiply the bill amount by the tip percentage:45.80 × 0.15 = 6.87]
[Add the tip to the original bill:
45.80 + 6.87 = 52.67 ]
[What I would tell the staff member:
“The tip amount is 6.87 and the total charge is 52.67.”]
