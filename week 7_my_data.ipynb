```python
# My Data Story: Example List
# Week 8 - Files and Friends


filename = "my_data.txt"


with open(filename, "r") as file:
    lines = file.readlines()

# Clean up each line (remove extra spaces and newlines)
lines = [line.strip() for line in lines]


print("== MY DATA STORY ==\n")

print("My data:")
for line in lines:
    print("-", line)

num_lines = len(lines)

longest_line = max(lines, key=len)

print("\n=== STATISTICS ===")
print("Number of lines:", num_lines)
print("Longest line:", longest_line)
```
