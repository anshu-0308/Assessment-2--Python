```python
def format_phone(phone):
    phone = str(phone)

    digits = ""
    for ch in phone:
        if ch.isdigit():
            digits += ch

    # if invalid, return empty (no error message)
    if len(digits) != 10:
        return ""

    return "(" + digits[:3] + ") " + digits[3:6] + "-" + digits[6:]

# Test cases
inputs = [
    "1234567890",
    "123-456-7890",
    "123 456 7890",
    "(123) 456-7890",
    "123456789",
    "abc-def-ghij",
    None
]

for i in inputs:
    result = format_phone(i)
    if result != "":
        print(result)
```
