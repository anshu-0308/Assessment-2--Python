**Scenario**: I need a function that converts a temperature value from Celsius to Fahrenheit.

**Specification:**

This function must convert a temperature value given in Celsius (°C) into its equivalent value in Fahrenheit (°F).

The temperature conversion function must:

•	Take a numeric input representing temperature in Celsius (°C).

•	Convert it to Fahrenheit using the standard conversion formula: 
      °F = (°C × 9/5) + 32
      
•	Return the converted Fahrenheit value 


**Examples:**  
•	0°C → 32°F (C=0)   {0°C = (0 × 9/5) + 32 = 32°F  
•	100°C → 212°F (boiling point of water)   
•	-40°C → -40°F (same value in both scales C = F) 



**Edge cases:**

•	Extreme low temperature (absolute limit case):

Input: -273.15°C → Output: -459.67°F
(This is approximately absolute zero, the lowest possible temperature in physics.) 



•	Very large value:
Input: 1,000,000°C → Output: 1,800,032°F 

________________________________________


# 1. Implementation with the help of AI 

**AI Reply:**

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Ask user for input
celsius = float(input("Enter temperature in Celsius: "))

# Convert and display result
fahrenheit = celsius_to_fahrenheit(celsius)

print("Temperature in Fahrenheit:", fahrenheit)
```
