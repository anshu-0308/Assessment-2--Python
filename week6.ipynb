```python
# Creating dictionary
grades = {
    "Math": 8,
    "Accounts": 11,
    "Programming": 4,
}

#add new subject
grades.update({"Science": 8})

grades.update({"Accounts": 12})

print("\nAfter change:")
for subject, grade in grades.items():
    print(f"{subject} : {grade}")


```

```python
grade = {
    "Programming": 8,
    "Math": 5,
    "Account": 3,
    "Science": 7
}

for subject, value in grade.items():
    print(subject, ":", value)
```
