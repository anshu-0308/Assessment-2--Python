# 2.	Add prints to trace execution

```python
def count_vowels(text):
    vowels = "aeiou"
    count = 0

    print("Input text:", text)

    for letter in text:
        print("Checking letter:", letter)

        # NO case conversion (as requested)
        if letter in vowels:
            count += 1
            print("Vowel found:", letter)
            print("Current count:", count)

    print("Final vowel count:", count)
    return count


# User input
text = input("Enter a word or sentence: ")

print("Number of vowels:", count_vowels(text))
```

# 5. Implement Simple Fix

Answer:

Fix 1: Added print for input → helps see what data is passed in

Fix 2: Added print for each letter → helps trace execution step-by-step

Fix 3 (Important): Used letter.lower() → fixes case-sensitivity issue (so uppercase vowels are counted)

Fix 4: Printed final count → helps confirm result before returning



```python
def count_vowels(text):
    vowels = "aeiou"
    count = 0

    # FIX 1: Print input for debugging
    print("Input text:", text)

    for letter in text:
        # FIX 2: Print each step of execution
        print("Checking letter:", letter)

        # FIX 3: Case-insensitive check
        if letter.lower() in vowels:
            count += 1
            print("Vowel found:", letter)
            print("Current count:", count)

    # FIX 4: Print final result before returning
    print("Final vowel count:", count)

    return count


# User input
text = input("Enter a word or sentence: ")

print("Number of vowels:", count_vowels(text))
```

The fix converts each letter to lowercase before checking if it is a vowel.
This allows the function to correctly count both uppercase and lowercase vowels, making the comparison case-insensitive.

**These tests verify:**
normal inputs, 
 uppercase inputs, 
 mixed-case inputs, 
 edge cases,

```python
def count_vowels(text):
    vowels = "aeiou"
    count = 0

    for letter in text:
        # Convert each letter to lowercase before checking
        if letter.lower() in vowels:
            count += 1

    return count


# TEST CASES

print('"Hello" =', count_vowels("Hello"))
print('"HELLO" =', count_vowels("HELLO"))
print('"Education" =', count_vowels("Education"))
print('"AEIOU" =', count_vowels("AEIOU"))
print('"Python" =', count_vowels("Python"))
print('"" =', count_vowels(""))



# ASSERTIONS


assert count_vowels("Hello") == 2
assert count_vowels("HELLO") == 2
assert count_vowels("Education") == 5
assert count_vowels("AEIOU") == 5
assert count_vowels("") == 0

print("\nAll test cases passed successfully.")
```
